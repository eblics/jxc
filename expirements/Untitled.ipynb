```python
import cv2 as cv
import matplotlib.pyplot as plt 
plt.figure(figsize=(10,10))
im = cv.imread('1.jpg')
imgray = cv.cvtColor(im, cv.COLOR_BGR2GRAY)
ret, thresh = cv.threshold(imgray, 127, 255, 0)
#plt.imshow(thresh,cmap='gray')
contours, hierarchy = cv.findContours(thresh, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)
#plt.imshow(im)
cv.drawContours(im, contours, 1001, (0,255,0), 3)
plt.imshow(im)
#print(contours[0])
```

```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
plt.figure(figsize=(20,10))
img = cv.imread('1.jpg',0)
ret,thresh = cv.threshold(img,160,255,cv.THRESH_BINARY)

edges = cv.Canny(thresh,0,100)
blur = cv.blur(edges,(5,5))


plt.subplot(151),plt.imshow(img,cmap = 'gray')
plt.subplot(152),plt.imshow(thresh,cmap = 'gray')
plt.subplot(153),plt.imshow(blur,cmap = 'gray')
plt.subplot(154),plt.imshow(edges,cmap = 'gray')

lines = cv.HoughLinesP(blur,1,np.pi/180,500,800)
ret,baw = cv.threshold(img,0,255,cv.THRESH_BINARY)
#print(lines)
for line in lines:
    x1,y1,x2,y2 = line[0]
    cv.line(baw,(x1,y1),(x2,y2),(0,0,0),2)
plt.subplot(155),plt.imshow(baw,cmap = 'gray')    
```
